```python
#a simple linear model where dependent variable is ‘MonthlyCharges’ and independent variable is ‘tenure’

def estimate_coefficients(x,y):
    n=x.count()
    mean_x, mean_y = x.mean(), y.mean()
    ss_xy = (y*x - n*mean_y*mean_x).sum()
    ss_xx = (x*x - n*mean_x*mean_x).sum()
    b1= ss_xy/ss_xx
    b0=mean_y - b1*mean_x
    return(b0,b1)

def plot_regression_line(x,y,b):
    fig = plt.figure(figsize=(10,10))
    plt.scatter(x,y)
    y_predicted=b[0] + b[1]*x
    plt.plot(x,y_predicted, color='red')
    plt.xlabel("tenure")
    plt.ylabel("Monthly Charges")
    plt.show()
    
x=cust_churn.tenure
y=cust_churn.MonthlyCharges

b=estimate_coefficients(x,y)
print("estimated coefficients: \nb={} \nb1={}".format(b[0], b[1]))
plot_regression_line(x,y,b)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
import numpy as np

#making x1 a 2D array
x1=pd.DataFrame(cust_churn['tenure'])
y1=cust_churn['MonthlyCharges']

#splitting the dataframe into train and test
x_train,x_test,y_train,y_test=train_test_split(x1,y1,test_size=0.3,random_state=0)

#Building the model
lm = LinearRegression().fit(x_train, y_train)

#Making prediction
y_predicted=lm.predict(x_test)
print("y-intercept",lm.intercept_)
print("coefficient:",lm.coef_)

#root mean squared error
mse = mse(y_predicted, y_test)
rmse= np.sqrt(mse)
print("Rmse:", rmse)

#mean absolute error
ma_error = mae(y_predicted, y_test)
print("MAE:", ma_error)

figure= plt.figure(figsize=(10,10))
plt.scatter(x_test, y_test)
plt.scatter(x_test,y_predicted)
```

```python
#logistic regression model where the dependent variable ‘Churn’ & independent variable is ‘MonthlyCharges’

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report


x_log= pd.DataFrame(cust_churn.MonthlyCharges)
y_log= cust_churn.Churn

x_train, x_test, y_train, y_test = train_test_split(x_log, y_log, test_size=0.35, random_state=1)
log_reg= LogisticRegression()
log_reg.fit(x_train, y_train)

y_pred=log_reg.predict(x_test)

y_pred = pd.Series(y_pred)
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))

```

```python
#a multiple logistic regression model where dependent variable is ‘Churn’ & independent variables are ‘tenure’ & ‘MonthlyCharges’

x = pd.DataFrame(cust_churn.loc[:,["tenure","MonthlyCharges"]])
y= cust_churn.Churn

x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
logmodel=LogisticRegression()
logmodel.fit(x_train, y_train)

y_pred=logmodel.predict(x_test)

print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))

```

```python
#a decision tree model where dependent variable is ‘Churn’ & independent variable is ‘tenure’

from sklearn.tree import DecisionTreeClassifier as dt

x=pd.DataFrame(cust_churn.tenure)
y=cust_churn.Churn

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=1)
tree= dt()
tree.fit(x_train, y_train)

y_pred=tree.predict(x_test)

print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
#a Random Forest model where dependent variable is ‘Churn’ & independent variables are ‘tenure’ and ‘MonthlyCharges’

from sklearn.ensemble import RandomForestClassifier 

x=cust_churn[["tenure","MonthlyCharges"]]
y=cust_churn.Churn

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1)
rf= RandomForestClassifier(n_estimators=100)
rf.fit(x_train, y_train)

y_pred= rf.predict(x_test)

print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test,y_pred))
```
